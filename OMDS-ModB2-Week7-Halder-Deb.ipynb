Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
# Supplier Market Share Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the data
suppliers = ['Supplier A', 'Supplier B', 'Supplier D', 'Supplier C']
market_shares = [34, 31, 26, 9]

# Create a DataFrame
df_market_share = pd.DataFrame({
    'Supplier': suppliers,
    'Market Share (%)': market_shares
})

# Sort the DataFrame by market share in descending order
df_market_share = df_market_share.sort_values('Market Share (%)', ascending=False)

# Set up the plot style
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Create a horizontal bar plot
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='Market Share (%)', y='Supplier', data=df_market_share,
                 palette='gray')

# Customize the plot
#plt.title('Supplier Market Share', fontsize=16, fontweight='bold')
#plt.xlabel('Market Share (%)', fontsize=12)
plt.ylabel('Supplier', fontsize=12)

# Add percentage labels to the end of each bar
for i, v in enumerate(df_market_share['Market Share (%)']):
    ax.text(v + 0.5, i, f'{v}%', va='center')

    # Remove grid, ticks, and box
ax.grid(False)
ax.set_xticks([])  # Remove X-axis ticks
ax.set_yticks([])  # Remove Y-axis ticks
ax.spines['top'].set_visible(False)    # Remove top border
ax.spines['right'].set_visible(False)  # Remove right border
ax.spines['bottom'].set_visible(False) # Remove bottom border
ax.spines['left'].set_visible(False)   # Remove left border

# Adjust layout and display total
plt.tight_layout()
plt.text(0.5, -0.1, 'Total: 100%', transform=plt.gca().transAxes,
         ha='center', fontweight='bold')
ax.grid(False)

# Display the plot
plt.show()

# Print the DataFrame
print(df_market_share)
```
